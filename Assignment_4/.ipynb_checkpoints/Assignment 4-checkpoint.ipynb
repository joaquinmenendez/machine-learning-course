```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.datasets import make_blobs
import pandas as pd
```

```python
###looking on the dataset
dataset = make_blobs(n_features=2, centers=2) #make_blob output is a tuple data [0], labels [1] 
#Ploting 
sns.scatterplot(x = dataset[0][:,0],y = dataset[0][:,1], hue = dataset[1])
plt.title('Make_blob 2 features')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
###KNN function
def knn_classif(dataset, centroid=2, plots = True, plot_final = True):
    ## Random assign start
    index = np.random.choice(dataset[0].shape[0], centroid, replace = False) 
    #chose N index as centroids (make sure that not selet the same index)
    centroids = dataset[0][index] #assign N random centroids
    #Ploting 
    colors = {9:'coral',8:'tomato',7:'magenta',6:'aqua',5:'khaki',
              4:'cyan',3:'orange',2:'green', 1:'red', 0:'blue'}
    if plots:
        plt.scatter(x = dataset[0][:,0],y = dataset[0][:,1])
        plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=200, c='k')
        plt.title('Initial Random centroids')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.show()
     
    ## Assign observation to the nearest mean
    #empty matrix (row is an observation, column the number of centroids)
    distances = np.zeros((dataset[0].shape[0],centroid))
    #calculate distances for all values to centroid
    for ncent in range(0,centroid):
        distn = np.sqrt(np.sum((dataset[0]-centroids[ncent])**2,axis = 1))
        distances[:,ncent] = distn
    #assign labels according to initial distances
    labels = np.argmin(distances, axis=1)
    if plots:
        plt.scatter(x = dataset[0][:,0],y = dataset[0][:,1], color = [colors[col] for col in labels], alpha= 0.4)
        plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=200, c=[colors[col] for col in range(0,centroid)])
        plt.title('First labels assigned')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.show()
    #recalculate centroid using labels (mean x, mean y)
    for cent in range(0,centroid):
        centroids[cent] = dataset[0][labels == cent].mean(axis = 0) #it calculates the centroid for each labeled class
    #Ploting  new initial centroids.
    if plots:
        plt.scatter(x = dataset[0][:,0],y = dataset[0][:,1], color = [colors[col] for col in labels], alpha= 0.4)
        plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=200, c=[colors[col] for col in range(0,centroid)])
        plt.title('Centroids initialized')
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.show()
    #This is going to be our initial condition. The loop is going to recalculate
    #new centrodis, and is going to re-label the datapoints until it converges.
    
    #calculate distance difference between previous centroid
    old_centroids = np.zeros(centroids.shape)
    cent_dist = np.linalg.norm(centroids - old_centroids, axis=None) # axis None output is only a number
    iteration = 0
    while cent_dist != 0: #threshold to stop the loop (convergence of centroids)
        old_centroids = np.copy(centroids) # pass the old centroids
        for ncent in range(0,centroid): #calculate distance over centroids
            distn = np.sqrt(np.sum((dataset[0]-centroids[ncent])**2,axis = 1))
            distances[:,ncent] = distn
        labels = np.argmin(distances, axis=1)
        #calculate new centroid
        for cent in range(0,centroid):
            centroids[cent] = dataset[0][labels == cent].mean(axis = 0) 
        iteration += 1
        cent_dist = np.linalg.norm(centroids - old_centroids, axis=None)
        if iteration > 500: break
    #Ploting  final clusters.
    if plot_final:
        plt.scatter(x = dataset[0][:,0],y = dataset[0][:,1], color = [colors[col] for col in labels], alpha= 0.4)
        plt.scatter(centroids[:,0], centroids[:,1], marker='*', s=200, c=[colors[col] for col in range(0,centroid)])    
        plt.title('Final clusters for K = %d' % centroid)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.show()
        print('Number of iterations: %d' % iteration)
    #calculate SSE
    SSE = 0    
    for K in range(0,centroid):
        l2 = np.linalg.norm((dataset[0][labels == K] - centroids[K]), axis=1)
        SSE = SSE + l2.sum()
    print('SSE for K-%d = %.2f' % (centroid,SSE))
    return SSE
```

```python
#KNN classif allow to plot the initial and final clusters
#an example:
plt.suptitle('Example of the KNN algorithm fitting a K=3 to a dataset with 2 centroids')
knn_classif(dataset, centroid=3, plot_final=True, plots = True)
```

```python
#Plotting SSE for different K_values
dataset = make_blobs(n_features=2, centers=2)
all_SSE = []
for k in range(1,11):
    all_SSE.append(knn_classif(dataset, centroid=k,plot_final=False, plots = False))
plt.plot(range(1,11),all_SSE)
plt.xticks(np.arange(10))
plt.xlabel('K values')
plt.ylabel('SSE')
plt.title('SSE~K-values for 2 cluster centers')
plt.show()
```

We can observe that when we applied the KNN algorithm to a 2 centers sample the rate of decrease of the SSE sharply shifts at K = 2. Usually the elbow method (looking where is the first of this shifts in the SSE decrease) can be used to roughly determine K.

```python
dataset = make_blobs(n_features=2, centers=5)
all_SSE = []
for k in range(1,11):
    all_SSE.append(knn_classif(dataset, centroid=k, plots = False, plot_final=False))
plt.plot(range(1,11),all_SSE)
plt.xticks(np.arange(10))
plt.xlabel('K values')
plt.ylabel('SSE')
plt.title('SSE~K-values for 5 cluster centers')
plt.show()
```

In the case of the sample with 5 centers we could observe a first 'elbow' at k = 2. Nevertheless, observing the SSE values we could observe that the 'stabilization' of this SSE decrease could be located around K=4 and K=5. It should be highlighted that in this case we are running only one simulation for each K, and this result are going to depend on the initial centroid assignment (a random process). In case of running several simulations and use the average SSE for each error we would see a shape more similar to the first graph, with a clear elbow in K = 5

(1) k-means, (2) agglomerative clustering, (3) Gaussian mixture models, (4) DBSCAN, and (5) spectral clustering

K-means. Is a  prototype-based, partitional clustering technique. This algorithm aims to partition the observations into k clusters in which each observation belongs to the cluster with the nearest mean (centroid). The data that is going to be clustered should allow calculating this centroid, in other words, the observations should be quantitative (level of measurement equal to interval or ratio) in order to estimate a meaningful mean. To cluster the data it starts randomly choosing n points (where n is the number of clusters defined for a user) among of all the observations and calculates the Euclidian distance from each point in the set to these randomly defined points. When all the observations have been assigned to a cluster, calculate the mean of the cluster(centroid). After this, the algorithm iterates until the observation's classes not change anymore (converge). K-means is sensitive to outliers and is recommended to remove then before clustering. K-means method cannot handle non-globular clusters or clusters of different sizes and densities. 

Agglomerative Clustering. It is a Hierarchial,Complete and Exclusive clustering algorithm. This algorithm starts with individual points as clusters,and successively merge the two closest clusters until only one cluster remains. To do so, it calculates a proximity matrix (different metrics, or linkages, could be applied) for all the observations and cluster on every iteration. Aglomerative clustering could be sensitive to outliers dependenig on the proximity measure used. But at the same time, it could isolate outliers (usually as singleton or small clusters that are not merging with other clusters).

Gaussian Mixture Models. Is a partitional clustering where each cluster is associated with a Gaussian distribution. Every observation is going to have a probability of belonging to all the clusters. In other words they are fuzzy clustering. The GMM use principally two parameters such as Mean and Variance for each cluster. It works well when there is variation in cluster variance and correlation between features, as well as when the clusters have equal variance.

DBSCAN. Is a density-based clustering algorithm that produces a partitional clustering. Unlike the previous techniques,  the number of clusters is automatically determined by the algorithm. The algorithm begins selecting an arbitrary observation as starting data point. All the observations within a certain space defined by the user, called epsilon, are going to be labeled as members of the clusters. Nevertheless, this is only going to happen if the number of observation within the epsilon space is equal or bigger to a minimum number of observations (this is also set by the user).
If there are a sufficient number of points then the clustering process starts and the current data point becomes the first point in the new cluster. Otherwise, the point will be labeled as noise. This procedure of making all points in the epsilon neighborhood belong to the same cluster is then repeated for all of the new points that have been just added to the cluster group. This process is repeated until all points in the cluster are determined. After this, a new observation is randomly selected and the process is repeated.DBSCAN has troubles when the clusters have very different densities or when the data is high-dimensional.
data

Spectral clustering. This technique first constructs an affinity matrix based on the observations, then it applies a dimensionality reduction on the data. To do so it uses the N first eigenvalues (selected by the number of cluster desired) of the similarity matrix in order to get the principal components of the data. Then calculates the similarity between data points in lower dimensions. It does not make many assumptions about the data, similar to k-means, allowing to produce good clustering results. Nevertheless is computationally slow for big datasets.

```python
aggregation = pd.read_table("./data/Aggregation.txt",delim_whitespace=True, names=["x1", "x2", "y"])
compound = pd.read_table("./data/Compound.txt",delim_whitespace=True, names=["x1", "x2", "y"])
d31 = pd.read_table("./data/D31.txt",delim_whitespace=True, names=["x1", "x2", "y"])
jain = pd.read_table("./data/jain.txt",delim_whitespace=True, names=["x1", "x2", "y"])
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
from sklearn.mixture import GaussianMixture

X_aggreg = np.ndarray((aggregation.shape[0],2)) 
X_aggreg[:,0] = aggregation.x1
X_aggreg[:,1] = aggregation.x2

X_compound = np.array(list(zip(compound.x1, compound.x2))).reshape(len(compound.x1), 2)
X_d31 = np.array(list(zip(d31.x1, d31.x2))).reshape(len(d31.x1), 2)
X_jain = np.array(list(zip(jain.x1, jain.x2))).reshape(len(jain.x1), 2)
```

```python
#plt.figure()
fig,ax = plt.subplots(ncols=5,nrows=4,figsize=(20,20),sharey=True)
plt.subplot(4,5,1)
plt.subplots_adjust(hspace = 0.4)
plt.suptitle('Fitting of 5 different cluster algorithms')

#1. K-means (aggregation data)
kmeans = KMeans(n_clusters=7)
kmeans.fit(X_aggreg)
y_kmeans = kmeans.predict(X_aggreg)
sns.scatterplot(x=aggregation.x1, y=aggregation.x2, hue=y_kmeans, palette="plasma",legend=False)
plt.title('K-means for Aggregation data')

#2. Agglomerative clustering(aggregation data)
plt.subplot(4,5,2)
cluster = AgglomerativeClustering(n_clusters=7,linkage='average').fit(X_aggreg)
aggregation["Classes"]= cluster.labels_
sns.scatterplot(x=aggregation.x1, y=aggregation.x2, hue=aggregation.Classes, palette="plasma",legend=False)
plt.title('Aggl. clustering for Aggregation data')

#3. GMM (aggregation data)
plt.subplot(4,5,3)
plt.title('GMM for Aggregation data')
gmm=GaussianMixture(n_components=7, covariance_type='full').fit(X_aggreg)
aggregation["Classes"]= gmm.predict(X_aggreg)
sns.scatterplot(x=aggregation.x1, y=aggregation.x2, hue=aggregation.Classes, palette="plasma",legend=False)

#4. DBSCAN (aggregation data)
plt.subplot(4,5,4)
plt.title('DBSCAN for Aggregation data')
#eps - max distance between points to be considered as the same class, min_samples - The number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=2, min_samples=14).fit(X_aggreg)
aggregation["Classes"]=dbscan.labels_
sns.scatterplot(x=aggregation.x1, y=aggregation.x2, hue=aggregation.Classes, palette="plasma", legend=False)

#5. Spectral clustering(aggregation data)
plt.subplot(4,5,5)
plt.title('Spectral clustering for Aggregation data')
spectral = SpectralClustering(n_clusters=7,
                assign_labels="discretize",
                random_state=0).fit(X_aggreg)
aggregation["Classes"]=spectral.labels_
sns.scatterplot(x=aggregation.x1, y=aggregation.x2, hue=aggregation.Classes, palette="plasma",legend='full')
plt.legend(bbox_to_anchor=(1.17, 1))

#6. K-means (compound data)
plt.subplot(4,5,6)
kmeans = KMeans(n_clusters=6)
kmeans.fit(X_compound)
y_kmeans = kmeans.predict(X_compound)
sns.scatterplot(x=compound.x1, y=compound.x2, hue=y_kmeans, palette="plasma",legend=False)
plt.title('K-means for compound data')
plt.ylim(5.0, 23)

#7. Agglomerative clustering (compound data)
plt.subplot(4,5,7)
plt.title('Aggl clustering for Compound data')
cluster = AgglomerativeClustering(n_clusters=6,linkage='ward').fit(X_compound)
compound["Classes"]= cluster.labels_
sns.scatterplot(x=compound.x1, y=compound.x2, hue=compound.Classes, palette="plasma",legend=False)

#8. GMM (compound data)
plt.subplot(4,5,8)
plt.title('GMM for Compound data')
gmm=GaussianMixture(n_components=6).fit(X_compound)
compound["Classes"]= gmm.predict(X_compound)
sns.scatterplot(x=compound.x1, y=compound.x2, hue=compound.Classes, palette="plasma",legend=False)

plt.subplot(4,5,9)
#9. DBSCAN (compound data)
plt.title('DBSCAN for Compound data')
dbscan = DBSCAN(eps=1).fit(X_compound)
compound["Classes"]=dbscan.labels_
sns.scatterplot(x=compound.x1, y=compound.x2, hue=compound.Classes, palette="plasma", legend=False)

#10. Spectral clustering(compound data)
plt.subplot(4,5,10)
plt.title('Spectral clustering for Compound data')
spectral = SpectralClustering(n_clusters=6,assign_labels="discretize",n_init=20).fit(X_compound)
compound["Classes"]=spectral.labels_
sns.scatterplot(x=compound.x1, y=compound.x2, hue=compound.Classes, palette="plasma",legend='full')
plt.legend(bbox_to_anchor=(1.17, 1))

#11.K-means (d31 data)
plt.subplot(4,5,11)
kmeans = KMeans(n_clusters=31)
kmeans.fit(X_d31)
y_kmeans = kmeans.predict(X_d31)
d31_kmeans = sns.scatterplot(x=d31.x1, y=d31.x2, hue=y_kmeans, palette="plasma", legend=False)
plt.title('K-means for D31 data')

#12. Agglomerative clustering (d31 data)
plt.subplot(4,5,12)
plt.title('Aggl. clustering for D31 data')
d31cluster = AgglomerativeClustering(n_clusters=31,linkage='average').fit(X_d31)
d31["Classes"]= d31cluster.labels_
sns.scatterplot(x=d31.x1, y=d31.x2, hue=d31.Classes, palette="plasma",legend=False)

#13. GMM (d31 data)
plt.subplot(4,5,13)
plt.title('GMM for D31 data')
gmm=GaussianMixture(n_components=31, covariance_type='full').fit(X_d31)
d31["Classes"]= gmm.predict(X_d31)
sns.scatterplot(x=d31.x1, y=d31.x2, hue=d31.Classes, palette="plasma",legend=False)

#14. DBSCAN (d31n data)
plt.subplot(4,5,14)
plt.title('DBSCAN for D31 data')
dbscan = DBSCAN(eps=.7, min_samples=20).fit(X_d31)
d31["Classes"]=dbscan.labels_
sns.scatterplot(x=d31.x1, y=d31.x2, hue=d31.Classes, palette="plasma", legend=False)

#15. Spectral clustering(d31 data)
plt.subplot(4,5,15)
plt.title('Spectral clustering for D31 data')
spectral = SpectralClustering(n_clusters=31,
                assign_labels="kmeans",
                random_state=0).fit(X_d31)
d31["Classes"]=spectral.labels_
sns.scatterplot(x=d31.x1, y=d31.x2, hue=d31.Classes, palette="plasma",legend='brief')
plt.legend(bbox_to_anchor=(1.7, 1))

#16. K-means (jain data)
plt.subplot(4,5,16)
kmeans = KMeans(n_clusters=2)
kmeans.fit(X_jain)
y_kmeans = kmeans.predict(X_jain)
jain_kmeans = sns.scatterplot(x=jain.x1, y=jain.x2, hue=y_kmeans, legend=False)
plt.title('K-means for jain data')


#17. Agglomerative clustering (jain data)
plt.subplot(4,5,17)
plt.title('Aggl. clustering for jain data')
cluster = AgglomerativeClustering(n_clusters=2,linkage='average').fit(X_jain)
jain["Classes"]= cluster.labels_
sns.scatterplot(x=jain.x1, y=jain.x2, hue=jain.Classes,legend=False)

#18. GMM (jain data)
plt.subplot(4,5,18)
plt.title('GMM for jain data')
gmm=GaussianMixture(n_components=2, covariance_type='tied').fit(X_jain)
jain["Classes"]= gmm.predict(X_jain)
sns.scatterplot(x=jain.x1, y=jain.x2, hue=jain.Classes,legend=False)

#19. DBSCAN (jain data)
plt.subplot(4,5,19)
plt.title('DBSCAN for jain data')
dbscan = DBSCAN(eps=2, min_samples=20).fit(X_jain)
jain["Classes"]=dbscan.labels_
color = {0:'steelblue', -1:'darkorange'}
sns.scatterplot(x=jain.x1, y=jain.x2, color= [color[col] for col in jain['Classes']], legend=False)

#20. Spectral clustering(jain data)
plt.subplot(4,5,20)
plt.title('Spectral clustering for jain data')
spectral = SpectralClustering(n_clusters=2,
                assign_labels="kmeans",
                random_state=0).fit(X_jain)
jain["Classes"]=spectral.labels_
sns.scatterplot(x=jain.x1, y=jain.x2, hue=jain.Classes,legend='full')
plt.legend(bbox_to_anchor=(1.6, 1))
```
